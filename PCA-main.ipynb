```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing
from scipy.ndimage.filters import gaussian_filter1d
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import colors


data = pd.read_excel('C:/Users/sony/Desktop/data/preprocess_5_feature.xlsx')

features = ['New Site','New Special Stock','New Time','Qty in unit of entry','holiday']
x = data.loc[:, features].values
print(type(x))
print('\n __________x________ \n',x)

x = StandardScaler().fit_transform(x)
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
print('\n Explained variation per principal component: {} \n'.format(pca.explained_variance_ratio_))

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])


fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection ='3d') 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

x = principalDf.loc[:,'principal component 1']
#print('x',x)
y = principalDf.loc[:,'principal component 2']
#print('y',y)
z = principalDf.loc[:,'principal component 3']
#print('z',z)

cmap = plt.cm.rainbow
norm = colors.BoundaryNorm(np.arange(-2.5, 3, 1), cmap.N)
ax.scatter3D(x, y, z, c=z, cmap=cmap, norm=norm)
ax.grid()
```
